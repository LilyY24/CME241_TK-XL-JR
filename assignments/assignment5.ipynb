## Question 1
You are a milkvendor and your task is to bring to your store a supply
(denoted $S \in \mathbb{R}$) of milk volume in the morning that will
give you the best profits. You know that the demand for milk through the
course of the day is a probability distribution function $f$ (for
mathematical convenience, assume people can buy milk in volumes that are
real numbers, hence milk demand $x \in \mathbb{R}$ is a continuous
variable with a probability density function). For every extra gallon of
milk you carry at the end of the day (supply $S$ exceeds random demand
$x$), you incur a cost of $h$ (effectively the wasteful purchases
amounting to the difference between your purchase price and end-of-day
discount disposal price since you are not allowed to sell the same milk
the next day). For every gallon of milk that a customer demands that you
don't carry (random demand $x$ exceeds supply $S$), you incur a cost of
$p$ (effectively the missed sales revenue amounting to the difference
between your sales price and purchase price). So your task is to
identify the optimal supply $S$ that minimizes your Expected Cost
$g(S)$, given by the following:

$$g_1(S) = E[\max(x-S, 0)] = \int_{-\infty}^{\infty} \max(x-S, 0) \cdot f(x) \cdot dx = \int_S^{\infty} (x-S) \cdot f(x) \cdot dx$$
$$g_2(S) = E[\max(S-x, 0)] = \int_{-\infty}^{\infty} \max(S-x, 0) \cdot f(x) \cdot dx = \int_{-\infty}^S (S-x) \cdot f(x) \cdot dx$$

$$g(S) = p \cdot g_1(S) + h \cdot g_2(S)$$

After you solve this problem, see if you can frame this problem in terms
of a call/put options portfolio problem.

Our expected total cost is given by:

$$
\begin{align*}
g(S) &= p \cdot g_1(S) + h \cdot g_2(S) \\
&= p \cdot \int_{S}^{\infty} (x-S) \cdot f(x) \cdot dx + h \cdot \int_{-\infty}^S (S-x) \cdot f(x) \cdot dx.
\end{align*}
$$

We calculate the derivative as:

$$
g'(S) = -p \cdot \int_S^{\infty} f(x) \cdot dx + h \cdot \int_{-\infty}^S f(x) \cdot dx,
$$

and set \( g'(S) = 0 \), then we get:

$$
\begin{align*}
&-p \cdot \int_S^{\infty} f(x) \cdot dx + h \cdot \int_{-\infty}^S f(x) \cdot dx = 0 \\
\text{then } \ \ &-p \cdot (1-F(S)) + h\cdot F(S) = 0 \\
\text{then } \ \ &F(S) = \frac{p}{p+h} \\
\text{so finally we get: } \ \ &S = F^{-1}\left(\frac{p}{p+h}\right).
\end{align*}
$$

Therefore, we obtain the optimal supply \( S \) as \( F^{-1}\left(\frac{p}{p+h}\right) \), where \( F \) is the cumulative distribution function of the demand of the milk.

Then we formulate this problem in terms of a call/put options portfolio as the following:

- Supply \( S \) corresponds to the strike price of the options
- Demand \( x \) corresponds to the spot price of the underlying asset in the financial market
- Overage cost \( h \) corresponds to the cost of holding a put option
- Underage cost \( p \) corresponds to the cost of writing a call option
- **Overage (Excess Supply)**: If the supply \( S \) exceeds the demand \( x \), the milk-vendor incurs a cost of \( h \) for every unsold gallon, similar to holding a put option that allows the vendor to sell milk at the "strike price" \( S \). If the "market price" \( x \) falls below \( S \), the vendor exercises the option and sells at \( S \), incurring a loss equivalent to the difference, which is the put option's payoff: \( \max(S - x, 0) \).
- **Underage (Unmet Demand)**: If the demand \( x \) exceeds the supply \( S \), the vendor incurs a cost of \( p \) for every gallon they can't sell, similar to the cost paid when the market price exceeds the strike price of a call option that the vendor has written (sold). The customer effectively "exercises" the option, and the vendor pays the difference: \( \max(x - S, 0) \).

Therefore, we can express the expected costs as the expected payoffs of these options:

- **Expected cost of overage \( g_1(S) \)** is akin to the expected payoff from holding put options: \( E[\max(S - x, 0)] \).
- **Expected cost of underage \( g_2(S) \)** is akin to the expected payoff from writing call options: \( E[\max(x - S, 0)] \).

To minimize the expected cost \( g(S) \), the vendor needs to choose a strike price \( S \) that balances the premium paid for the put options and the premium received from writing the call options. This optimal \( S \) is where the combined cost (or the net premium of the options portfolio) is minimized.

Assume an arbitrary probability distribution for asset price movements from one time step to another, so you'd be sampling from the arbitrary
transition probability distribution. Hence, you will be solving this problem with Approximate Dynamic Programming using the code in

```python
import numpy as np
from typing import Iterable, Tuple, Callable
from rl.distribution import SampledDistribution, Choose
from rl.markov_decision_process import MarkovDecisionProcess
from rl.markov_process import MarkovProcess
from rl.markov_process import NonTerminal, State
from rl.approximate_dynamic_programming import evaluate_mrp, ValueFunctionApprox, NTStateDistribution
from rl.function_approx import Tabular
from rl.policy import Policy
from rl.distribution import Distribution

StateType = float
ActionType = float

class AssetPriceMDP(MarkovDecisionProcess[float, float]):
    def __init__(self, sigma: float, mu: float, dt: float):
        self.sigma = sigma
        self.mu = mu
        self.dt = dt

    def actions(self, price: NonTerminal[float]) -> Iterable[float]:
        return [-1, 0, 1]

    def step(self, price: NonTerminal[float], action: float) -> SampledDistribution[Tuple[State[float], float]]:
        def sample_next_state_and_reward() -> Tuple[State[float], float]:
            next_price = price.state * np.exp((self.mu - 0.5 * self.sigma**2) * self.dt + self.sigma * np.sqrt(self.dt) * np.random.normal())
            reward = action * (next_price - price.state)
            return (NonTerminal(next_price), reward)

        return SampledDistribution(sample_next_state_and_reward)
class SimplePolicy(Policy[StateType, ActionType]):
    def act(self, state: NonTerminal[StateType]) -> Distribution[ActionType]:
        return Choose([-1, 0, 1])



mdp = AssetPriceMDP(sigma=0.2, mu=0.05, dt=1/252)
approx_0: ValueFunctionApprox[StateType] = Tabular()
policy = SimplePolicy()
# policy = lambda _: Choose([-1, 0, 1])
non_terminal_states_distribution: NTStateDistribution[StateType] = Choose([NonTerminal(s) for s in np.linspace(50, 150, 100)])
gamma = 0.99

value_func_approx_iter = evaluate_mrp(
    mrp=mdp.apply_policy(policy),
    γ=gamma,
    approx_0=approx_0,
    non_terminal_states_distribution=non_terminal_states_distribution,
    num_state_samples=100
)

for _ in range(1000):
    approx = next(value_func_approx_iter)

final_approx: ValueFunctionApprox[StateType] = approx
```

We'd like to build a simple simulator of Order Book Dynamics as a
`MarkovProcess` using the code in
[rl/chapter9/order_book.py](https://github.com/TikhonJelvis/RL-book/blob/master/rl/chapter9/order_book.py).
An object of type `OrderBook` constitutes the *State*. Your task is to
come up with a simple model for random arrivals of Market Orders and
Limit Orders based on the current contents of the `OrderBook`. This
model of random arrivals of Marker Orders and Limit Orders defines the
probabilistic transitions from the current state (`OrderBook` object) to
the next state (`OrderBook` object). Implement the probabilistic
transitions as a `MarkovProcess` and use it's `simulate` method to
complete your implementation of a simple simulator of Order Book
Dynamics.

Experiment with different models for random arrivals of Market Orders
and Limit Orders.

**Solution**
We implemented the Order Book Dynamics using two forms of random arrival processes:

1. There's a 50% chance of receiving a market order and a 50% chance of receiving a limit order. For market orders, there's an equal chance of it being a buy or sell order. For limit orders, the price is randomly chosen to be within a range above the current ask (for sell limit orders) or below the current bid (for buy limit orders), and the quantity will be randomly generated.

2. Market Order and Limit Order arrives following a Poisson Process, each controlled by an independent parameter (entered as function argument). The price is randomly chosen to be within a range above the current ask (for sell limit orders) or below the current bid (for buy limit orders), and the quantity will be randomly generated.

```python
from rl.distribution import Categorical
from typing import Iterable
from rl.chapter9.order_book import OrderBook, DollarsAndShares
from rl.markov_process import MarkovProcess, Terminal, NonTerminal, Distribution, State

import numpy as np
from typing import Tuple


class OrderBookMarkovProcess(MarkovProcess[OrderBook]):

    def __init__(self):
        pass

    def transition(self, state: NonTerminal[OrderBook]) -> Distribution[State[OrderBook]]:
        if np.random.rand() < 0.5:
            # Market Order
            is_buy = np.random.rand() < 0.5
            quantity = np.random.randint(1, 100)
            if is_buy:
                _, new_state = self.hashable_to_orderbook(state.state).buy_market_order(quantity)
            else:
                _, new_state = self.hashable_to_orderbook(state.state).sell_market_order(quantity)
        else:
            # Limit Order
            is_buy = np.random.rand() < 0.5
            quantity = np.random.randint(1, 100)
            price_shift = np.random.uniform(-5, 5)
            if is_buy:
                price = max(1, self.hashable_to_orderbook(state.state).bid_price() + price_shift)
                _, new_state = self.hashable_to_orderbook(state.state).buy_limit_order(price, quantity)
            else:
                price = self.hashable_to_orderbook(state.state).ask_price() + price_shift
                _, new_state = self.hashable_to_orderbook(state.state).sell_limit_order(price, quantity)

        # return Categorical({NonTerminal(new_state): 1.0})

        hashable_state = self.orderbook_to_hashable(new_state)
        return Categorical({NonTerminal(hashable_state): 1.0})

    def orderbook_to_hashable(self, orderbook: OrderBook) -> Tuple:
        # Convert the order book's bids and asks into a hashable tuple format
        bids_tuple = tuple((d_s.dollars, d_s.shares) for d_s in orderbook.descending_bids)
        asks_tuple = tuple((d_s.dollars, d_s.shares) for d_s in orderbook.ascending_asks)
        return (bids_tuple, asks_tuple)

    def hashable_to_orderbook(self, orderbook_tuple) -> OrderBook:
        bids_tuple, asks_tuple = orderbook_tuple

        # Convert each tuple of (dollars, shares) back into DollarsAndShares objects
        descending_bids = []
        for d_s in bids_tuple:
          if isinstance(d_s, Tuple):
            dollars, shares = d_s
            descending_bids.append(DollarsAndShares(dollars=dollars, shares=shares))
          else:
            descending_bids.append(d_s)

        ascending_asks = []
        for d_s in asks_tuple:
          if isinstance(d_s, Tuple):
            dollars, shares = d_s
            ascending_asks.append(DollarsAndShares(dollars=dollars, shares=shares))
          else:
            ascending_asks.append(d_s)

        orderbook = OrderBook(descending_bids=descending_bids, ascending_asks=ascending_asks)

        return orderbook

    def simulate(
        self,
        start_state_distribution: Distribution[NonTerminal[OrderBook]]
    ) -> Iterable[State[OrderBook]]:
        return super().simulate(start_state_distribution)


if __name__ == "__main__":
    # Initialize the order book with some bids and asks
    initial_order_book = OrderBook(descending_bids=(DollarsAndShares(dollars=1.0, shares=20), DollarsAndShares(dollars=0.8, shares=50), DollarsAndShares(dollars=0.7, shares=10)),
                                   ascending_asks=(DollarsAndShares(dollars=0.4, shares=30), DollarsAndShares(dollars=0.6, shares=10)))

    # Define a start state distribution
    start_state_dist = Categorical({NonTerminal((initial_order_book.descending_bids, initial_order_book.ascending_asks)): 1.0})

    # Create an instance of the OrderBookMarkovProcess
    ob_mp = OrderBookMarkovProcess()

    # Simulate the order book dynamics
    for state in ob_mp.simulate(start_state_dist):
        if len(state.state[0]) == 0 or len(state.state[1]) == 0:
          break
        print(state)
```

```python
import numpy as np
from numpy.random import poisson

class PoissonOrderBookMarkovProcess(MarkovProcess[OrderBook]):

    def __init__(self, lambda_limit_order: float, lambda_market_order: float):
        self.lambda_limit_order = lambda_limit_order
        self.lambda_market_order = lambda_market_order

    def transition(self, state: NonTerminal[OrderBook]) -> Distribution[State[OrderBook]]:
        if np.random.rand() < 0.5:
            # Market Order
            is_buy = np.random.rand() < 0.5
            # Generate quantity based on Poisson distribution
            quantity = poisson(self.lambda_market_order)
            if is_buy:
                _, new_state = self.hashable_to_orderbook(state.state).buy_market_order(quantity)
            else:
                _, new_state = self.hashable_to_orderbook(state.state).sell_market_order(quantity)
        else:
            # Limit Order
            is_buy = np.random.rand() < 0.5
            # Generate quantity based on Poisson distribution
            quantity = poisson(self.lambda_limit_order)
            price_shift = np.random.uniform(-5, 5)
            if is_buy:
                price = max(1, self.hashable_to_orderbook(state.state).bid_price() + price_shift)
                _, new_state = self.hashable_to_orderbook(state.state).buy_limit_order(price, quantity)
            else:
                price = self.hashable_to_orderbook(state.state).ask_price() + price_shift
                _, new_state = self.hashable_to_orderbook(state.state).sell_limit_order(price, quantity)

        # return Categorical({NonTerminal(new_state): 1.0})
        hashable_state = self.orderbook_to_hashable(new_state)

        # Use hashable_state in the Categorical distribution
        return Categorical({NonTerminal(hashable_state): 1.0})

    def orderbook_to_hashable(self, orderbook: OrderBook) -> Tuple:
        # Convert the order book's bids and asks into a hashable tuple format
        bids_tuple = tuple((d_s.dollars, d_s.shares) for d_s in orderbook.descending_bids)
        asks_tuple = tuple((d_s.dollars, d_s.shares) for d_s in orderbook.ascending_asks)
        return (bids_tuple, asks_tuple)

    def hashable_to_orderbook(self, orderbook_tuple) -> OrderBook:
        bids_tuple, asks_tuple = orderbook_tuple

        # Convert each tuple of (dollars, shares) back into DollarsAndShares objects
        descending_bids = []
        for d_s in bids_tuple:
          if isinstance(d_s, Tuple):
            dollars, shares = d_s
            descending_bids.append(DollarsAndShares(dollars=dollars, shares=shares))
          else:
            descending_bids.append(d_s)

        ascending_asks = []
        for d_s in asks_tuple:
          if isinstance(d_s, Tuple):
            dollars, shares = d_s
            ascending_asks.append(DollarsAndShares(dollars=dollars, shares=shares))
          else:
            ascending_asks.append(d_s)

        # Create a new OrderBook instance with the reconstructed lists
        orderbook = OrderBook(descending_bids=descending_bids, ascending_asks=ascending_asks)

        return orderbook


if __name__ == "__main__":
    # Initialize the order book with some bids and asks
    initial_order_book = OrderBook(descending_bids=(DollarsAndShares(dollars=1.0, shares=20), DollarsAndShares(dollars=0.8, shares=50), DollarsAndShares(dollars=0.7, shares=10)),
                                   ascending_asks=(DollarsAndShares(dollars=0.4, shares=30), DollarsAndShares(dollars=0.6, shares=10)))

    # Define a start state distribution
    start_state_dist = Categorical({NonTerminal((initial_order_book.descending_bids, initial_order_book.ascending_asks)): 1.0})

    # Create an instance of the OrderBookMarkovProcess
    ob_mp = PoissonOrderBookMarkovProcess(lambda_limit_order=0.1, lambda_market_order=0.2)

    # Simulate the order book dynamics
    for state in ob_mp.simulate(start_state_dist):
        if len(state.state[0]) == 0 or len(state.state[1]) == 0:
          break
        print(state)
```
